```python
%matplotlib inline
import math
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**2 + 2 * math.cos(2*x)

a = 0
b = 3
delta = 0.125

# --- Ընտրանքների եղանակ ---
x = a
min_x = x
min_f = f(x)

x_vals = []
y_vals = []

while x <= b:
    fx = f(x)
    x_vals.append(x)
    y_vals.append(fx)

    if fx < min_f:
        min_f = fx
        min_x = x

    x += delta

print("Մինիմումը մոտավորապես գտնվում է x =", min_x)
print("f(x) =", min_f)

# --- Գրաֆիկ ---
x_plot = np.linspace(a, b, 400)
y_plot = x_plot**2 + 2 * np.cos(2*x_plot)

plt.figure()
plt.plot(x_plot, y_plot, label="f(x)")
plt.scatter(x_vals, y_vals, label="Քայլեր (δ)")
plt.scatter(min_x, min_f, label="Մինիմում")

plt.legend()
plt.grid()
plt.show()
plt.savefig("graph.png")
```
