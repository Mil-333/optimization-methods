```
import math
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**2 + 2 * math.cos(2*x)

a = 0
b = 3
delta = 0.125

# --- Ընտրանքների եղանակ ---
x = a
min_x = x
min_f = f(x)

x_vals = []
y_vals = []

print(f"{'Քայլ': <6} | {'x': <10} | {'f(x)': <10}")
print("-" * 30)

step_count = 0
while x <= b:
    fx = f(x)
    x_vals.append(x)
    y_vals.append(fx)

    # Տպում ենք ընթացիկ քայլը
    print(f"{step_count: <6} | {x: <10.4f} | {fx: <10.4f}")

    if fx < min_f:
        min_f = fx
        min_x = x

    x += delta
    step_count += 1

print("-" * 30)
print(f"Մինիմումը մոտավորապես գտնվում է x = {min_x}")
print(f"f(x) = {min_f}")

# --- Գրաֆիկ ---
x_plot = np.linspace(a, b, 400)
y_plot = x_plot**2 + 2 * np.cos(2 * x_plot)

plt.figure(figsize=(10, 6))
plt.plot(x_plot, y_plot, label="f(x)", color='blue', alpha=0.5)
plt.scatter(x_vals, y_vals, color='red', s=20, label="Դիտարկված կետեր")
plt.scatter(min_x, min_f, color='green', s=100, label="Գտնված մինիմում", zorder=5)

plt.title("Մինիմումի որոնում քայլ առ քայլ")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 0.25 * x**4 + x**2 - 8*x + 12

# Smooth curve
x = np.linspace(0, 2, 200)
y = f(x)

# Uniform points
x_points = np.array([0, 0.5, 1.0, 1.5, 2.0])
y_points = f(x_points)

plt.figure()
plt.plot(x, y, label="f(x)")
plt.scatter(x_points, y_points, color='red', label="Search Nodes (x_k)")

for i in range(len(x_points)):
    plt.text(x_points[i], y_points[i] + 0.3, f"{y_points[i]:.4f}", ha='center')

plt.title("Method of Alternatives: Uniform Search Grid")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.legend()

plt.savefig("optimization_plot.png")
plt.show()
```
